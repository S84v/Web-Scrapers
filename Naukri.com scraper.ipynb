```python
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import Select
from selenium.common.exceptions import NoSuchElementException
from selenium import webdriver
import pandas as pd
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.support.ui import WebDriverWait

url = 'https://www.naukri.com/data-analyst-jobs'
driver = webdriver.Chrome()
driver.get(url)
driver.maximize_window()
driver.implicitly_wait(5)
accept_btn = driver.find_element(By.XPATH, '/html/body/div[1]/div[4]/div[2]/div/button')
accept_btn.click()
job_lst = []
c = 0

idx = 1
jobs = driver.find_elements(by = By.CLASS_NAME, value = 'jobTuple')

while c < 15:
    for job in jobs:
        try:
            print("try block", idx)
            name = driver.find_element(By.XPATH,'/html/body/div[1]/div[4]/div/div/section[2]/div[2]/article['+str(idx)+']/div[1]/div/a').text
            comp = driver.find_element(By.XPATH,'/html/body/div[1]/div[4]/div/div/section[2]/div[2]/article['+str(idx)+']/div[1]/div[1]/div/a[1]').text
            loc = driver.find_element(By.XPATH,'/html/body/div[1]/div[4]/div/div/section[2]/div[2]/article['+str(idx)+']/div[1]/ul/li[3]/span').text
            exp = driver.find_element(By.XPATH,'/html/body/div[1]/div[4]/div/div/section[2]/div[2]/article['+str(idx)+']/div[1]/ul/li[1]/span[1]').text
            ctc = driver.find_element(By.XPATH,'/html/body/div[1]/div[4]/div/div/section[2]/div[2]/article['+str(idx)+']/div[1]/ul/li[2]/span[1]').text
            link = driver.find_element(By.XPATH,'/html/body/div[1]/div[4]/div/div/section[2]/div[2]/article['+str(idx)+']/div[1]/div/a')
            l = link.get_attribute('href')

            job_dict = {
                    "Name": name,
                    "Company": comp,
                    "Location": loc,
                    "Experience": exp,
                    "CTC": ctc,
                    "Job Link": l
                    }

            job_lst.append(job_dict)
            idx += 1

        except NoSuchElementException:
            print("catch block", c)
            nxt_page = driver.find_element(By.XPATH, '/html/body/div[1]/div[4]/div/div/section[2]/div[3]/div/a[2]')
            nxt_page.click()
            idx = 1
            c += 1
            break
            

df = pd.DataFrame(job_lst)

df.to_csv("C:\\Users\\a839995\\Desktop\\jupyter\\Naukri.csv", index=False)
```
